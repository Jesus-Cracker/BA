# `00_validate.ipynb` — End-to-end validation of the contactless AF pipeline

**What this notebook does**

1. **Validate the ground truth (GT).** Detect R-peaks on the *reference* ECG with a robust,
   artifact-aware detector (neurokit2), build the RR series per window, and check *when / where*
   AF is present — then **measure the detector against the known clinical labels** so we can trust
   it instead of eyeballing. Includes overlay plots (ECG + detected peaks) as direct proof.
2. **Extract features** for the three contactless modalities (cECG, PPG, BCG); BCG uses the
   paper-faithful **SWT+PAPR** morphology features (no RR/HRV — there is no reliable BCG peak
   detector, which is exactly why RR is dropped for BCG).
3. **Train three experts** (cECG; PPG = ppg1+ppg2; BCG = bcg1+bcg2), each on its own signals,
   under **Leave-One-Patient-Out (LOPO)**.
4. **SQI gating network** — a small NN that reads per-window signal-quality (SQI) and outputs 3
   weights to blend the experts.
5. **Results** — per-expert + gated metrics, **plus a per-window inspector** (expert guesses and
   gate weights for any single window) and a **diagnostics** section explaining *why* the gate
   does or doesn't help, with concrete improvement levers.

The reference ECG is used **only** to make/validate labels — never as an expert feature (at
deployment there is no reference ECG; the point is *contactless* detection).

> **All knobs live in Section 0.** A consolidated "what to try" list is in the final section.
> **Future:** the `GatingNet` is written to be swapped for a PyTorch module + Optuna once the
> design is shown to beat the naive mean here.

## 2 · Validate the ground truth — *robustly*

**Why the previous detector failed.** A homemade derivative→square→threshold detector is fragile:
on a noisy 30 s sinus window it finds far too many peaks (spurious) and on a motion/amplitude
artifact it finds too few. Either way the RR series looks irregular → the window is wrongly called
**AF**. That is exactly the pattern of false-AF on non-AF patients you saw.

**Fix:** detect R-peaks with **neurokit2** (adaptive thresholding + artifact correction), run **once
over the whole recording** (more context = better thresholds), then slice the RR series per window.
A per-window **quality gate** marks windows with implausible HR / too few beats as *n/a* instead of
forcing an AF decision, so bad detection can never masquerade as AF.

We then **measure the detector against your clinical labels** (AF list) rather than trusting it
blindly: AF patients should be mostly AF-windows, controls mostly non-AF.

```python
def detect_rpeaks(ecg_filt, fs, method=RPEAK_METHOD):
    """R-peak sample indices over a full recording. neurokit2 if available, else basic fallback."""
    x = np.asarray(ecg_filt, float)
    x = np.where(np.isfinite(x), x, np.nanmedian(x[np.isfinite(x)]) if np.isfinite(x).any() else 0.0)
    if HAVE_NK:
        try:
            clean = nk.ecg_clean(x, sampling_rate=fs, method=method)
            _, info = nk.ecg_peaks(clean, sampling_rate=fs, method=method, correct_artifacts=True)
            return np.asarray(info["ECG_R_Peaks"], int)
        except Exception as e:
            print("neurokit failed, fallback:", e)
    # ---- basic fallback (Pan–Tompkins-lite) ----
    d = np.diff(x, prepend=x[0]); sq = d * d
    w = max(1, int(round(0.150 * fs)))
    integ = np.convolve(sq, np.ones(w) / w, mode="same")
    thr = 0.30 * (np.mean(integ) + np.std(integ))
    pk, _ = find_peaks(integ, height=thr, distance=max(1, int(round(0.20 * fs))))
    return pk


def rr_af_metrics(rr):
    """AF-relevant RR-irregularity metrics for one window."""
    drr = np.diff(rr)
    h, _ = np.histogram(drr, bins=16, range=(-0.4, 0.4))
    p = h / max(h.sum(), 1); p = p[p > 0]
    return dict(n_rr=len(rr), mean_hr=60.0 / np.mean(rr),
                rmssd=float(np.sqrt(np.mean(drr ** 2))),
                cv=float(np.std(rr) / np.mean(rr)),
                pnn=float(np.mean(np.abs(drr) > 0.05)),
                shannon=float(-np.sum(p * np.log(p)) / np.log(16)))


def classify_af(m, cv_thr=AF_CV_THR, pnn_thr=AF_PNN_THR):
    if m is None: return np.nan                       # undecidable window
    return int((m["cv"] >= cv_thr) or (m["pnn"] >= pnn_thr))


def window_metrics(rpeaks_full, fs, win_start, win_len):
    """RR metrics for the window [win_start, win_start+win_len] samples, with a quality gate."""
    pk = rpeaks_full[(rpeaks_full >= win_start) & (rpeaks_full < win_start + win_len)]
    if len(pk) < GT_MIN_BEATS: return None
    rr = np.diff(pk) / fs
    rr = rr[(rr > RR_PHYS[0]) & (rr < RR_PHYS[1])]
    if len(rr) < GT_MIN_BEATS - 1: return None
    m = rr_af_metrics(rr)
    if not (GT_HR_RANGE[0] <= m["mean_hr"] <= GT_HR_RANGE[1]): return None   # implausible -> n/a
    return m


def gt_window_table(gt_filt, gt_fs, window_s):
    """Non-overlapping per-window GT table for one patient (detection done once on full record)."""
    rpk = detect_rpeaks(gt_filt, gt_fs)
    n, wlen = len(gt_filt), int(window_s * gt_fs)
    rows = []
    for k, start in enumerate(range(0, max(1, n - wlen + 1), wlen)):
        m = window_metrics(rpk, gt_fs, start, wlen)
        rows.append(dict(win_idx=k, t_start_s=start / gt_fs, label=classify_af(m),
                         **(m or dict(n_rr=0, mean_hr=np.nan, rmssd=np.nan,
                                      cv=np.nan, pnn=np.nan, shannon=np.nan))))
    return pd.DataFrame(rows), rpk
```

```python
# Run GT validation for every patient.
gt_tables, gt_rpeaks, gt_signals, summary = {}, {}, {}, []
for pid in patients:
    try:
        pat = PatientSignals(str(Path(DATA_ROOT) / pid))
        pat.filter_all(FS); pat.offset_correction()      # GT gets only alignment.json flips
        if pat.gt_ecg_filt is None:
            print(f"{pid}: no GT ECG, skipped"); continue
        g = np.asarray(pat.gt_ecg_filt, float)
        t, rpk = gt_window_table(g, GT_FS, WINDOW_SEC)
        gt_tables[pid], gt_rpeaks[pid], gt_signals[pid] = t, rpk, g
        dec = t["label"].dropna()
        summary.append(dict(patient=pid, clinical=("AF" if pid in af_set else "non-AF"),
                            n_win=len(t), n_decided=int(dec.size),
                            af_frac=float((dec == 1).mean()) if dec.size else np.nan,
                            median_hr=float(np.nanmedian(t["mean_hr"]))))
    except Exception as e:
        print(f"{pid}: GT validation failed -> {e}")

gt_summary = pd.DataFrame(summary)
if not gt_summary.empty:
    display(gt_summary.sort_values(["clinical", "af_frac"], ascending=[True, False])
                      .round(3).reset_index(drop=True))
```

### Data-completeness audit

The PPG experts compute RR/HRV features via heartpy, which legitimately fails on some noisy
windows (those features become NaN). **No window is dropped** — `HistGradientBoostingClassifier`
consumes NaN natively — but it's worth seeing *how much* is missing per modality so "lost data" is
a measured number, not a guess. (The console warnings that used to accompany this came from heartpy
running inside joblib workers; they're now scoped-suppressed in `features.py` without hiding any
real failure.) If a modality's RR features are missing on a large fraction of windows, the
established fix is to compute them with the robust neurokit2 detector you already use for the GT,
instead of heartpy.

```python
audit = []
for m in ["cecg", "ppg", "bcg"]:
    cols = X.expert_feature_cols(lab, m)
    sub = lab[cols]
    rr_cols = [c for c in cols if any(k in c.lower() for k in
               ("meanrr","sdnn","rmssd","pnn","sd1","sd2","cosen","dfa","shannon","turning"))]
    audit.append(dict(modality=m, n_feature_cols=len(cols),
                      mean_nan_frac=round(float(sub.isna().mean().mean()), 4),
                      windows_all_nan=int((sub.isna().all(axis=1)).sum()),
                      rr_window_missing=(round(float(lab[rr_cols].isna().any(axis=1).mean()), 4)
                                         if rr_cols else 0.0)))
display(pd.DataFrame(audit))
print("mean_nan_frac = avg fraction of NaN cells; windows_all_nan = windows with NO usable feature\n"
      "for that modality; rr_window_missing = fraction of windows missing >=1 RR/HRV feature.")
```

## 4 · Three modality experts (LOPO out-of-fold probabilities)

One `HistGradientBoostingClassifier` per modality (handles NaN windows natively; swap freely),
each restricted to its own columns. Every window's probability comes from a model that never saw
that patient — these OOF probabilities feed the gate and every metric below.

```python
def metrics(y, p, thr=0.5):
    y = np.asarray(y).astype(int); pred = (np.asarray(p) >= thr).astype(int)
    tn, fp, fn, tp = confusion_matrix(y, pred, labels=[0, 1]).ravel()
    g = lambda a, b: a / b if b else np.nan
    return dict(acc=(tp+tn)/len(y), sens=g(tp, tp+fn), spec=g(tn, tn+fp),
                ppv=g(tp, tp+fp), npv=g(tn, tn+fn), f1=g(2*tp, 2*tp+fp+fn),
                auc=roc_auc_score(y, p) if len(np.unique(y)) == 2 else np.nan)

MODALITIES = ["cecg", "ppg", "bcg"]
y = lab["y"].astype(int).values
groups = lab["patient"].values

def expert_oof_probs(data, y, groups):
    P = np.full((len(data), len(MODALITIES)), np.nan)
    for j, mod in enumerate(MODALITIES):
        Xm = data[X.expert_feature_cols(data, mod)].to_numpy(float)
        for tr, te in LeaveOneGroupOut().split(Xm, y, groups):
            if len(np.unique(y[tr])) < 2:
                P[te, j] = y[tr].mean(); continue
            clf = HistGradientBoostingClassifier(random_state=RANDOM_STATE)
            clf.fit(Xm[tr], y[tr], sample_weight=compute_sample_weight("balanced", y[tr]))
            P[te, j] = clf.predict_proba(Xm[te])[:, 1]
    return P

P = expert_oof_probs(lab, y, groups)
expert_df = pd.DataFrame({m: metrics(y, P[:, j]) for j, m in enumerate(MODALITIES)}).T.round(3)
display(expert_df)
```

## 5 · SQI gating network

The gate maps per-window **SQI** → softmax over the three experts (how much to trust each
modality this window); final probability `p̂ = Σ wᵢ·pᵢ`, supervised by the labels (BCE).

Two robustness features matter (set in Section 0): a **softmax temperature** and **weight decay**
that bias the gate toward *equal* weights, so in the worst case the gate degrades gracefully to the
naive mean instead of overfitting SQI noise. (SQI alone is ~chance for AF — its only job is
*routing*, not classifying.)

```python
class GatingNet:
    """SQI -> softmax(weights over experts); blends expert probs. 1 hidden tanh layer, BCE loss.
       Near-uniform init + temperature + weight decay => graceful fallback to the mean."""
    def __init__(self, d_in, n_experts, hidden=GATE_HIDDEN, lr=GATE_LR, epochs=GATE_EPOCHS,
                 temp=GATE_TEMP, wd=GATE_WD, seed=RANDOM_STATE):
        r = np.random.default_rng(seed)
        self.W1 = r.normal(0, 1/np.sqrt(d_in), (d_in, hidden)); self.b1 = np.zeros(hidden)
        self.W2 = r.normal(0, 0.01, (hidden, n_experts));        self.b2 = np.zeros(n_experts)
        self.lr, self.epochs, self.temp, self.wd = lr, epochs, temp, wd
        self.mu = self.sd = None

    def _softmax(self, z):
        z = (z / self.temp); z = z - z.max(1, keepdims=True)
        e = np.exp(z); return e / e.sum(1, keepdims=True)

    def weights(self, Xsqi):
        Xs = (Xsqi - self.mu) / self.sd
        return self._softmax(np.tanh(Xs @ self.W1 + self.b1) @ self.W2 + self.b2)

    def fit(self, Xsqi, P, y):
        self.mu, self.sd = Xsqi.mean(0), Xsqi.std(0) + 1e-8
        Xs = (Xsqi - self.mu) / self.sd; y = y.astype(float); eps = 1e-7
        for _ in range(self.epochs):
            h = np.tanh(Xs @ self.W1 + self.b1)
            w = self._softmax(h @ self.W2 + self.b2)
            phat = np.clip((w * P).sum(1), eps, 1 - eps)
            dphat = (phat - y) / (phat * (1 - phat)) / len(y)
            dz2 = (dphat[:, None] * w * (P - phat[:, None])) / self.temp
            dh = (dz2 @ self.W2.T) * (1 - h ** 2)
            self.W2 -= self.lr * (h.T @ dz2 + self.wd * self.W2); self.b2 -= self.lr * dz2.sum(0)
            self.W1 -= self.lr * (Xs.T @ dh + self.wd * self.W1); self.b1 -= self.lr * dh.sum(0)
        return self

    def predict_proba(self, Xsqi, P):
        return (self.weights(Xsqi) * P).sum(1)


sqi_cols = X.gate_sqi_cols(lab, "all")
Xsqi = np.nan_to_num(lab[sqi_cols].to_numpy(float))

gated = np.full(len(lab), np.nan); sqi_only = np.full(len(lab), np.nan)
for tr, te in LeaveOneGroupOut().split(Xsqi, y, groups):
    if len(np.unique(y[tr])) < 2:
        gated[te] = P[tr].mean(); sqi_only[te] = y[tr].mean(); continue
    gated[te] = GatingNet(Xsqi.shape[1], len(MODALITIES)).fit(Xsqi[tr], P[tr], y[tr]) \
                    .predict_proba(Xsqi[te], P[te])
    sqi_only[te] = LogisticRegression(max_iter=2000, class_weight="balanced") \
                       .fit(Xsqi[tr], y[tr]).predict_proba(Xsqi[te])[:, 1]

print(f"SQI-only AUC (sanity, expect ~0.5): {roc_auc_score(y, sqi_only):.3f}")
print(f"Naive mean-of-experts AUC:          {roc_auc_score(y, P.mean(1)):.3f}")
print(f"SQI-GATED (plain) experts AUC:      {roc_auc_score(y, gated):.3f}")
```

## 5b · Reliability gate — a better NN training objective

Same network shape, **different goal**. Instead of training only on the blended-probability loss
(which only sees disagreement windows), the reliability gate also learns a **per-expert routing
target**: from the SQI, predict how reliable each expert is on this window — the exact signal
Section 8d found in the SQI. Loss = `α·routing-BCE + (1−α)·blend-BCE + β·(negative weight-entropy)`.
Dense supervision on every window makes it sample-efficient (it matters: your independent unit is
the patient, ~40), and it stops the gate from landing *below* the mean.

This is still a NN producing the 3 weights (what your supervisor wants); only the objective and the
penalties change. `GATE_ALPHA / GATE_BETA / GATE_SOFT` are in Section 0.

```python
class ReliabilityGate:
    """SQI -> 3 logits z.  weights = softmax(z/temp);  reliability r = sigmoid(z).
       Loss = alpha*BCE(r, expert_correct) + (1-alpha)*BCE(blend, y) + beta*sum_k w_k log w_k.
       (beta>0 encourages decisive weights.)  Same shape as GatingNet -> easy PyTorch port."""
    def __init__(self, d_in, n_experts, hidden=GATE_HIDDEN, lr=GATE_LR, epochs=GATE_EPOCHS,
                 temp=GATE_TEMP, wd=1e-4, alpha=GATE_ALPHA, beta=GATE_BETA, soft=GATE_SOFT,
                 seed=RANDOM_STATE):
        r = np.random.default_rng(seed)
        self.W1 = r.normal(0, 1/np.sqrt(d_in), (d_in, hidden)); self.b1 = np.zeros(hidden)
        self.W2 = r.normal(0, 0.05, (hidden, n_experts));        self.b2 = np.zeros(n_experts)
        self.lr, self.epochs, self.temp, self.wd = lr, epochs, temp, wd
        self.alpha, self.beta, self.soft = alpha, beta, soft; self.mu = self.sd = None

    def _sm(self, z):
        z = z / self.temp; z = z - z.max(1, keepdims=True); e = np.exp(z); return e / e.sum(1, keepdims=True)

    def weights(self, Xsqi):
        Xs = (Xsqi - self.mu) / self.sd
        return self._sm(np.tanh(Xs @ self.W1 + self.b1) @ self.W2 + self.b2)

    def fit(self, Xsqi, P, y):
        self.mu, self.sd = Xsqi.mean(0), Xsqi.std(0) + 1e-8
        Xs = (Xsqi - self.mu) / self.sd; y = y.astype(float); eps = 1e-7; N = len(y); K = P.shape[1]
        T = np.where(y[:, None] == 1, P, 1 - P) if self.soft else \
            ((P >= .5).astype(int) == y[:, None]).astype(float)
        for _ in range(self.epochs):
            h = np.tanh(Xs @ self.W1 + self.b1); z = h @ self.W2 + self.b2
            w = self._sm(z); r = 1 / (1 + np.exp(-z))
            ph = np.clip((w * P).sum(1), eps, 1 - eps)
            dz_blend = ((ph - y) / (ph * (1 - ph)) / N)[:, None] * w * (P - ph[:, None]) / self.temp
            dz_route = (r - T) / (N * K)
            wl = np.log(w + eps); dz_ent = (self.beta / self.temp) * w * (wl - (w * wl).sum(1, keepdims=True)) / N
            dz = self.alpha * dz_route + (1 - self.alpha) * dz_blend + dz_ent
            dh = (dz @ self.W2.T) * (1 - h ** 2)
            self.W2 -= self.lr * (h.T @ dz + self.wd * self.W2); self.b2 -= self.lr * dz.sum(0)
            self.W1 -= self.lr * (Xs.T @ dh + self.wd * self.W1); self.b1 -= self.lr * dh.sum(0)
        return self

    def predict_proba(self, Xsqi, P):
        return (self.weights(Xsqi) * P).sum(1)


gated_rel = np.full(len(lab), np.nan)
for tr, te in LeaveOneGroupOut().split(Xsqi, y, groups):
    if len(np.unique(y[tr])) < 2:
        gated_rel[te] = P[tr].mean(); continue
    gated_rel[te] = ReliabilityGate(Xsqi.shape[1], len(MODALITIES)).fit(Xsqi[tr], P[tr], y[tr]) \
                        .predict_proba(Xsqi[te], P[te])

print(f"Naive mean-of-experts AUC:     {roc_auc_score(y, P.mean(1)):.3f}")
print(f"plain gate AUC:                {roc_auc_score(y, gated):.3f}")
print(f"RELIABILITY gate AUC:          {roc_auc_score(y, gated_rel):.3f}  "
      f"(alpha={GATE_ALPHA}, beta={GATE_BETA})")
```

## 5c · Error-prediction fusion (Bachelet pipeline, adapted to AF)

This is the closest possible port of Marc Bachelet's HR-fusion design to your classification task.
The pipeline is **decoupled**, exactly like his:

1. A fully-connected NN takes the **rich multimodal features** (all modality feature columns
   jointly, ~150 inputs, like his 155) and **predicts each expert's error** on the window.
   Because your experts output probabilities, the error target is the Brier/abs error
   `|pₖ − y|` (his was `|HFₖ − HF_ref|`). It's a plain multi-output regression — a dense,
   well-posed signal on every window for every expert (his predicted-vs-true error reached r≈0.89).
2. A **fixed decay** maps predicted error → weight: `w = 1` for `err ≤ e₀`, else `exp(−(err−e₀)/τ)`,
   then normalised. Small predicted error → high weight (his `e₀, τ`).
3. **Weighted average** of the calibrated expert probabilities.

The NN never sees the fused output during training, and the experts are fixed upstream — same as
his. Everything is nested in LOPO + per-fold isotonic calibration so the number is honest. The one
deliberate departure (the *different goal*): error is `|p−y|` on a bounded probability, not an
unbounded HR error — so expect a smaller lift than his, for the structural reasons we discussed.

```python
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

def errpred_input(data):
    if ERRPRED_INPUT == "sqi":
        cols = X.gate_sqi_cols(data, "all")
    elif ERRPRED_INPUT == "features":
        cols = sum((X.expert_feature_cols(data, m) for m in MODALITIES), [])
    else:  # both
        cols = sum((X.expert_feature_cols(data, m) for m in MODALITIES), []) + X.gate_sqi_cols(data, "all")
    return data[cols].to_numpy(float)

def err_to_weight(E, e0=DECAY_E0, tau=DECAY_TAU):
    w = np.where(E <= e0, 1.0, np.exp(-(E - e0) / tau))
    s = w.sum(1, keepdims=True); s[s == 0] = 1.0
    return w / s

Xin = errpred_input(lab)
gated_err = np.full(len(lab), np.nan)
pred_err_oof = np.full((len(lab), len(MODALITIES)), np.nan)   # for the diagnostic below

for tr, te in LeaveOneGroupOut().split(Xin, y, groups):
    if len(np.unique(y[tr])) < 2:
        gated_err[te] = P[tr].mean(); continue
    Ptr, Pte = P[tr].copy(), P[te].copy()
    if ERRPRED_CALIBRATE:                                     # nested isotonic calibration
        for j in range(len(MODALITIES)):
            iso = IsotonicRegression(out_of_bounds="clip").fit(Ptr[:, j], y[tr])
            Ptr[:, j] = iso.predict(Ptr[:, j]); Pte[:, j] = iso.predict(Pte[:, j])
    # per-expert error target on the (calibrated) train probabilities
    Etr = (Ptr - y[tr][:, None]) ** 2 if ERRPRED_TARGET == "brier" else np.abs(Ptr - y[tr][:, None])
    reg = make_pipeline(SimpleImputer(strategy="median"), StandardScaler(),
                        MLPRegressor(hidden_layer_sizes=ERRPRED_HIDDEN, activation="relu",
                                     solver="adam", alpha=ERRPRED_ALPHA, max_iter=300,
                                     early_stopping=True, n_iter_no_change=15,
                                     random_state=RANDOM_STATE))
    reg.fit(Xin[tr], Etr)
    Ete = np.clip(reg.predict(Xin[te]), 0, 1)
    pred_err_oof[te] = Ete
    gated_err[te] = (err_to_weight(Ete) * Pte).sum(1)

# how well does the NN predict each expert's error? (his r=0.89 benchmark)
true_err = np.abs(P - y[:, None])
print("predicted-vs-true error correlation per expert (his benchmark r=0.89):")
for j, m in enumerate(MODALITIES):
    ok = np.isfinite(pred_err_oof[:, j])
    r = np.corrcoef(pred_err_oof[ok, j], true_err[ok, j])[0, 1]
    print(f"  {m}: r={r:.3f}")
print(f"\nNaive mean AUC          : {roc_auc_score(y, P.mean(1)):.3f}")
print(f"reliability gate AUC    : {roc_auc_score(y, gated_rel):.3f}")
print(f"ERROR-PRED fusion AUC   : {roc_auc_score(y, gated_err):.3f}   "
      f"(input={ERRPRED_INPUT}, target={ERRPRED_TARGET}, e0={DECAY_E0}, tau={DECAY_TAU})")
```

## 5d · Optuna tuning of the error-prediction fusion (optional)

Set `RUN_OPTUNA = True` in Section 0 (needs `pip install optuna`). Each trial is scored on an
**inner** grouped validation split held out *within* the training patients — never the LOPO test
patient — so the search can't overfit the evaluation, exactly as Bachelet did. It tunes the NN
(hidden sizes, L2, learning rate) and the decay (`e₀, τ`) jointly against fused-AUC.

```python
if RUN_OPTUNA:
    import optuna
    from sklearn.model_selection import GroupShuffleSplit
    optuna.logging.set_verbosity(optuna.logging.WARNING)

    def objective(trial):
        hid = trial.suggest_categorical("hidden", [(64,), (128, 64), (256, 128)])
        alpha = trial.suggest_float("alpha", 1e-5, 1e-2, log=True)
        lr = trial.suggest_float("lr", 1e-4, 5e-3, log=True)
        e0 = trial.suggest_float("e0", 0.02, 0.30)
        tau = trial.suggest_float("tau", 0.05, 0.60)
        aucs = []
        gss = GroupShuffleSplit(n_splits=3, test_size=0.3, random_state=RANDOM_STATE)
        for itr, ival in gss.split(Xin, y, groups):     # inner split on TRAIN-like data only
            if len(np.unique(y[ival])) < 2: continue
            Ptr2, Pval = P[itr].copy(), P[ival].copy()
            for j in range(len(MODALITIES)):
                iso = IsotonicRegression(out_of_bounds="clip").fit(Ptr2[:, j], y[itr])
                Ptr2[:, j] = iso.predict(Ptr2[:, j]); Pval[:, j] = iso.predict(Pval[:, j])
            Etr = np.abs(Ptr2 - y[itr][:, None])
            reg = make_pipeline(SimpleImputer(strategy="median"), StandardScaler(),
                                MLPRegressor(hidden_layer_sizes=hid, alpha=alpha,
                                             learning_rate_init=lr, max_iter=300,
                                             early_stopping=True, random_state=RANDOM_STATE))
            reg.fit(Xin[itr], Etr)
            Ev = np.clip(reg.predict(Xin[ival]), 0, 1)
            blended = (err_to_weight(Ev, e0, tau) * Pval).sum(1)
            aucs.append(roc_auc_score(y[ival], blended))
        return np.mean(aucs) if aucs else 0.5

    study = optuna.create_study(direction="maximize")
    study.optimize(objective, n_trials=OPTUNA_TRIALS, show_progress_bar=False)
    print("best inner-validation AUC:", round(study.best_value, 3))
    print("best params:", study.best_params)
    print("\nPlug best params into Section 0 (ERRPRED_HIDDEN/ALPHA, DECAY_E0/TAU) and rerun 5c "
          "for the honest LOPO number.")
else:
    print("Optuna disabled (RUN_OPTUNA=False). Enable in Section 0 to tune.")
```

## 7 · Per-window inspector — expert guesses + gate weights

For any single window: the three expert AF-probabilities, the gate's weights, the blended
probability, the truth, and the SQI values that drove the weights. Use it to see *why* a window is
right or wrong — e.g. an expert is confidently wrong but the gate failed to down-weight it.

```python
W_all = g_all.weights(Xsqi)                    # gate weights for every labelled window
gated_all = (W_all * P).sum(1)

def inspect_window(pid=None, win_idx=None, row=None):
    """Inspect one window by (patient, win_idx) or by integer row index into `lab`."""
    if row is None:
        sel = lab.index[(lab.patient == pid) & (lab.win_idx == win_idx)]
        if len(sel) == 0: print("no such window"); return
        row = int(sel[0])
    r = lab.iloc[row]
    print(f"patient={r.patient}  win_idx={int(r.win_idx)}  t_start={r.t_start_s:.0f}s  "
          f"TRUE={'AF' if r.y==1 else 'non-AF'}")
    print(f"  expert probs : " + "  ".join(f"{m}={P[row,j]:.2f}" for j,m in enumerate(MODALITIES)))
    print(f"  gate weights : " + "  ".join(f"{m}={W_all[row,j]:.2f}" for j,m in enumerate(MODALITIES)))
    print(f"  blended p-hat: {gated_all[row]:.2f}  -> predict {'AF' if gated_all[row]>=.5 else 'non-AF'}")
    comp = [c for c in sqi_cols if c.endswith("_composite")]
    if comp:
        print("  SQI(composite): " + "  ".join(f"{c.replace('sqi_','').replace('_composite','')}="
              f"{lab.iloc[row][c]:.2f}" for c in comp))
    fig, ax = plt.subplots(1, 2, figsize=(8, 2.6))
    ax[0].bar(MODALITIES, P[row], color="#26a69a"); ax[0].set_ylim(0,1)
    ax[0].axhline(.5, ls="--", c="grey"); ax[0].set_title("expert P(AF)")
    ax[1].bar(MODALITIES, W_all[row], color="#5c6bc0"); ax[1].set_ylim(0,1)
    ax[1].set_title("gate weights"); plt.tight_layout(); plt.show()

# Example: the most "wrong" window (largest gap between blended prediction and truth)
worst = int(np.argmax(np.abs(gated_all - y)))
print("=== Example: worst-predicted window ===")
inspect_window(row=worst)

def list_disagreements(n=10):
    """Windows where experts disagree most (std of their probs) — prime debugging candidates."""
    d = lab.copy()
    d["expert_std"] = P.std(1); d["blended"] = gated_all; d["truth"] = y
    cols = ["patient","win_idx","truth","blended","expert_std"]
    return d.sort_values("expert_std", ascending=False)[cols].head(n).round(3)

print("\n=== Windows where experts disagree most ===")
display(list_disagreements())
# Inspect any of them, e.g.:  inspect_window("PAT001", 12)
```

## 8 · Diagnostics — *why* the gate helps (or doesn't), and how to fix it

If the gated AUC is **below** the naive mean, the gate is adding variance without useful routing.
The cells below quantify whether useful routing even exists and try concrete fixes.

```python
# ── 8a · Routing headroom: is there anything for a gate to exploit? ───────────
mean_auc = roc_auc_score(y, P.mean(1))
conf_toward_truth = np.where(y[:, None] == 1, P, 1 - P)     # how "right" each expert is
oracle = P[np.arange(len(P)), conf_toward_truth.argmax(1)]
oracle_auc = roc_auc_score(y, oracle)
mean_wrong = (P.mean(1) >= .5).astype(int) != y
any_expert_right = (((P >= .5).astype(int) == y[:, None]).any(1))
rescuable = (mean_wrong & any_expert_right).mean()
print(f"naive mean AUC          : {mean_auc:.3f}")
print(f"SQI-gated AUC           : {roc_auc_score(y, gated):.3f}")
print(f"oracle (upper bound) AUC: {oracle_auc:.3f}")
print(f"windows the mean gets wrong but some expert gets right: {rescuable:.1%}")
print("\nWHAT THE ORACLE IS: for each window it picks the expert that is most right — using the\n"
      "TRUE label to choose. So it PEEKS AT THE ANSWER. No gate can reach it, because at inference\n"
      "the gate never sees the label. The oracle is a ceiling, not a target.\n"
      "The honest headroom is the line above: the % of windows a PERFECT router could flip. The\n"
      "windows where all experts agree and are all wrong (like the inspector example) are\n"
      "unreachable by ANY weighting. Whether a real gate can capture even part of the headroom\n"
      "depends on whether its inputs predict which expert is right — tested in 8d.")
```

```python
# ── 8b · Calibrate expert probabilities, then blend (often helps the ensemble) ─
Pc = np.zeros_like(P)
for j in range(len(MODALITIES)):
    iso = IsotonicRegression(out_of_bounds="clip").fit(P[:, j], y)
    Pc[:, j] = iso.predict(P[:, j])
print(f"mean-of-experts AUC  raw={roc_auc_score(y, P.mean(1)):.3f}  "
      f"calibrated={roc_auc_score(y, Pc.mean(1)):.3f}")
print("(diagnostic: isotonic fit in-sample on OOF probs; for a final number nest it in the LOPO loop)")
```

```python
# ── 8c · STREAMING early-detection (the deployment goal: flag AF fast) ─────────
# Per recording, walk windows in time order; flag AF the first time STREAM_K of the last
# STREAM_M windows exceed STREAM_TAU. Report: detection rate on AF patients, median latency
# to first flag, and the false-alarm rate on controls. This is what a 1-5 min device cares
# about — early, persistent detection — not a single per-window AUC.
def streaming_eval(score, K=STREAM_K, M=STREAM_M, tau=STREAM_TAU):
    d = lab[["patient", "t_start_s"]].copy(); d["y"] = y; d["s"] = score
    det_rate = {"AF": [], "non-AF": []}; latencies = []; fp = 0; n_ctrl = 0
    for pid, g in d.sort_values("t_start_s").groupby("patient"):
        is_af = int(g["y"].iloc[0] == 1); over = (g["s"].values >= tau).astype(int)
        flagged_at = None
        for i in range(len(over)):
            lo = max(0, i - M + 1)
            if over[lo:i + 1].sum() >= K:
                flagged_at = g["t_start_s"].values[i]; break
        grp = "AF" if is_af else "non-AF"
        det_rate[grp].append(int(flagged_at is not None))
        if is_af and flagged_at is not None: latencies.append(flagged_at)
        if not is_af:
            n_ctrl += 1; fp += int(flagged_at is not None)
    print(f"rule: {K}-of-{M} windows >= {tau}")
    print(f"  AF detection rate   : {np.mean(det_rate['AF']):.2f}  ({sum(det_rate['AF'])}/{len(det_rate['AF'])} patients)")
    print(f"  median time-to-flag : {np.median(latencies):.0f}s  (AF patients that were detected)")
    print(f"  control false-alarm : {fp}/{n_ctrl} patients flagged")

print(">> plain gate"); streaming_eval(gated)
print(">> reliability gate"); streaming_eval(gated_rel)
print(">> error-pred fusion"); streaming_eval(gated_err)
print("\nTweak STREAM_K / STREAM_M / STREAM_TAU in Section 0 to trade latency against false alarms.\n"
      "Higher K/M and higher tau = fewer false alarms but slower detection.")
```

```python
# ── 8d · Is per-window expert correctness PREDICTABLE? (the key oracle question) ─
# A gate can only approach the oracle if its inputs predict which expert is right on a
# window. Test it directly: target_k = 1 if expert k is correct on the window; train a
# LOPO classifier from the gate inputs (SQI) to predict it. AUC ~0.5 => SQI cannot route,
# so a pure SQI gate is capped near the mean (explains gate <= mean). AUC >0.5 => routable.
print("Can the gate's inputs predict which expert is correct? (LOPO AUC, 0.5 = no signal)")
for src_name, src in [("SQI only", Xsqi),
                      ("SQI + expert probs", np.hstack([Xsqi, P]))]:
    aucs = []
    for j in range(len(MODALITIES)):
        correct = ((P[:, j] >= .5).astype(int) == y).astype(int)
        if len(np.unique(correct)) < 2:
            aucs.append(np.nan); continue
        oof = np.full(len(lab), np.nan)
        for tr, te in LeaveOneGroupOut().split(src, correct, groups):
            if len(np.unique(correct[tr])) < 2:
                oof[te] = correct[tr].mean(); continue
            m = LogisticRegression(max_iter=2000, class_weight="balanced").fit(src[tr], correct[tr])
            oof[te] = m.predict_proba(src[te])[:, 1]
        aucs.append(roc_auc_score(correct, oof))
    print(f"  {src_name:20s}: " + "  ".join(f"{m}={a:.3f}" for m, a in zip(MODALITIES, aucs)))
print("If 'SQI only' ~0.5 but 'SQI + expert probs' >0.5, the routing signal lives in the\n"
      "experts' agreement/confidence, not in SQI -> use a STACKING meta-learner (8e), not SQI gating.")
```

```python
# ── 8e · STACKING meta-learner: the realistic way to capture oracle headroom ───
# Instead of SQI->weights, learn final_prob = f(expert probs + their agreement [+ SQI]).
# Fully nested in LOPO: experts' OOF probs are calibrated on train patients, the meta-model
# is trained on train patients and evaluated on the held-out patient. This is the honest
# number to compare against the naive mean.
USE_SQI_IN_META = False     # try True to let the meta-model also see SQI

def meta_features(Pmat, Xs=None):
    f = [Pmat,
         Pmat.mean(1, keepdims=True), Pmat.std(1, keepdims=True),
         Pmat.max(1, keepdims=True), Pmat.min(1, keepdims=True),
         np.abs(Pmat[:, [0]] - Pmat[:, [1]]),
         np.abs(Pmat[:, [0]] - Pmat[:, [2]]),
         np.abs(Pmat[:, [1]] - Pmat[:, [2]])]
    if Xs is not None:
        f.append(Xs)
    return np.hstack(f)

stack_oof = np.full(len(lab), np.nan)
for tr, te in LeaveOneGroupOut().split(P, y, groups):
    if len(np.unique(y[tr])) < 2:
        stack_oof[te] = P[tr].mean(); continue
    Ptr, Pte = P[tr].copy(), P[te].copy()
    for j in range(len(MODALITIES)):                      # nested isotonic calibration
        iso = IsotonicRegression(out_of_bounds="clip").fit(Ptr[:, j], y[tr])
        Ptr[:, j] = iso.predict(Ptr[:, j]); Pte[:, j] = iso.predict(Pte[:, j])
    Xtr = Xsqi[tr] if USE_SQI_IN_META else None
    Xte = Xsqi[te] if USE_SQI_IN_META else None
    Ftr, Fte = meta_features(Ptr, Xtr), meta_features(Pte, Xte)
    meta = LogisticRegression(max_iter=3000, class_weight="balanced").fit(Ftr, y[tr])
    stack_oof[te] = meta.predict_proba(Fte)[:, 1]

print(f"naive mean AUC      : {roc_auc_score(y, P.mean(1)):.3f}")
print(f"SQI-gated AUC       : {roc_auc_score(y, gated):.3f}")
print(f"STACKED meta AUC    : {roc_auc_score(y, stack_oof):.3f}   (nested calibration, LOPO)")
display(pd.DataFrame({"mean(experts)": metrics(y, P.mean(1)),
                      "SQI-gated":     metrics(y, gated),
                      "stacked-meta":  metrics(y, stack_oof)}).T.round(3))
```

```python
# ── 8f · Patient-level bootstrap CIs — is any difference real, or just noise? ──
# Resample PATIENTS (not windows) to respect the real unit of independence (n=40).
rng_bs = np.random.default_rng(RANDOM_STATE)
pat_ids = lab["patient"].values
uniq = np.array(sorted(set(pat_ids)))

def auc_for(scores, chosen):
    mask = np.isin(pat_ids, chosen)
    yy = y[mask]
    return roc_auc_score(yy, scores[mask]) if len(np.unique(yy)) == 2 else np.nan

def boot_ci(scores, n=1000):
    vals = []
    for _ in range(n):
        chosen = rng_bs.choice(uniq, size=len(uniq), replace=True)
        a = auc_for(scores, chosen)
        if not np.isnan(a): vals.append(a)
    return np.nanpercentile(vals, [2.5, 50, 97.5])

for name, sc in [("mean(experts)", P.mean(1)), ("SQI-gated (plain)", gated),
                 ("SQI-gated (reliab.)", gated_rel), ("error-pred fusion", gated_err),
                 ("stacked-meta", stack_oof)]:
    lo, mid, hi = boot_ci(sc)
    print(f"{name:16s} AUC median={mid:.3f}  95% CI [{lo:.3f}, {hi:.3f}]")
print("\nIf the CIs overlap heavily (they will), 'mean vs gated vs stacked' differences are within\n"
      "noise at n=40 — pick the SIMPLEST model that is not worse, and report the CI in the thesis.")
```

## 9 · What to try / change

**Excluding patients** — add ids to `EXCLUDE_PATIENTS` in Section 0. They are dropped from the GT
check, features, experts and gate. Use it to park the ambiguous-GT patients until your supervisor
meeting; nothing is excluded automatically.

**Error-prediction fusion (Section 5c — the Bachelet port)** — the recommended combiner. It feeds
the rich multimodal features (`ERRPRED_INPUT`), predicts each expert's error (`ERRPRED_TARGET`),
maps error→weight via a decay (`DECAY_E0`, `DECAY_TAU`), and blends calibrated experts — decoupled,
nested, LOPO-honest. Watch the printed predicted-vs-true error correlation per expert: that is the
analog of his r≈0.89 and tells you directly how much routing signal exists. Tune the NN and decay
with `RUN_OPTUNA=True` (Section 5d), then copy the best params back into Section 0.

**Improving the combiner (read Section 8 in order)** — 8a shows the *oracle* ceiling but it peeks at
labels, so it is not a target. 8d tests whether per-window expert correctness is predictable from
the gate inputs; if SQI scores >0.5 there (it does for you), the routing signal exists and the job
is to train the gate to use it. The **reliability gate (Section 5b)** is the fix: a NN with the same
3-weight output but trained on a per-expert routing target (`GATE_ALPHA`) plus optional decisiveness
penalty (`GATE_BETA`), so it learns from every window instead of only disagreement windows and stops
landing below the mean. The **8e stacking meta-learner** is an alternative combiner. 8f puts
bootstrap CIs on all of them; at n=40 patients expect heavy overlap — so the realistic goal is a NN
that reliably *matches or slightly beats* the mean (which justifies its use), not a large jump. The
weak routing signal (0.66-0.72) and small patient count, not the architecture, set the ceiling.

**Detection goal (Section 8c)** — your device must flag AF *early*. Tune `STREAM_K/M/TAU` to trade
detection latency against control false-alarms; report detection rate + median time-to-flag +
false-alarm rate. This window-streaming view, not a per-window AUC, is the number that matches the
1-5 minute use-case.

**Window length** — `WINDOW_SEC` (10–30). Shorter windows give more *rows* but NOT more independent
data: the unit of independence is the patient (40), not the window. Expect little generalisation
gain and noisier wavelet/frequency features (FFT bin width = 1/T, so 10 s triples the bin size and
RR/irregularity stats get shakier). Worth a quick sensitivity check, but it will not unlock the
oracle — the oracle gap is about *routing*, not window count.

**Optuna** — useful, but later. It tunes hyperparameters; it adds no information. Run it only after
8d/8e show there is headroom your inputs can capture, and always with nested CV (an inner validation
split inside each LOPO fold) or you will overfit the search at n=40. Tuning the experts + calibration
will likely matter more than tuning the gate.

**GT validation (Section 2)**
- `RPEAK_METHOD` — `'neurokit'`, `'pantompkins1985'`, `'hamilton2002'`. Compare beat counts in the
  overlay plot; pick whatever tracks the ECG best on your hardware.
- `GT_MIN_BEATS`, `GT_HR_RANGE` — the quality gate. Tighten if noisy windows still slip through as
  AF; loosen if too many windows go *n/a*.
- `AF_CV_THR`, `AF_PNN_THR` — how irregular RR must be to call AF. Use the QA scatter/histogram to
  pick a threshold that separates your AF and control patients.
- The QA cell flags patients whose detector AF-fraction contradicts their clinical label — inspect
  those recordings directly (paroxysmal AF, lead noise, or a label issue).

**Labels** — `LABEL_SOURCE='patient'` for your 21-AF / 19-control cohort. `'gt_window'` only if a
cohort is all-AF.

**Experts** — swap `HistGradientBoostingClassifier` for RandomForest/SVM/etc; keep balanced
`sample_weight`. BCG wavelet hyperparameters (`db4`, level 6, PAPR domain) live in `features.py`.

**Gate** — `GATE_TEMP` (up = softer, closer to equal weights), `GATE_WD` (up = stronger pull to the
mean), `GATE_HIDDEN/EPOCHS/LR`. Read Section 8a first: only invest in the gate if the oracle shows
routing headroom. Otherwise the **naive mean is the honest baseline to beat**.

**Inspector (Section 7)** — `inspect_window("PAT0xx", k)` for any window; `list_disagreements()` to
find the informative ones.

**Decision level** — report **patient-level** metrics (Section 8c) as the headline; per-window as a
secondary, finer-grained view.

**PyTorch + Optuna (future)** — replace `GatingNet` with an `nn.Module` of the same shape
(`forward(SQI) -> softmax(3)`, `p̂ = Σ wᵢpᵢ`, BCE). Keep the LOPO loop. Let Optuna tune
hidden size/depth/lr/weight-decay/temperature. Move here once Section 8a shows the gate can beat the
mean — the current results suggest tuning the experts and calibration may matter more than the gate.

**Always keep LOPO.** Windows of one patient are highly correlated; mixing them across train/test
inflates every score.
